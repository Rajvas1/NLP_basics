```python
def wish(name):   # outer function
    greet = "hey" # outer function variable

    def make_message():   # Inner function
        return f' {greet} , {name}'
    
    message = make_message()

    print(message)

```

```python
wish("rajashekhar")
```

```python
def greet(call):
    greeting = "hey"

    def message_do():
        return f" {greeting} , {call}"
    ask = message_do()

    print(ask)
    
```

```python
greet("who are you")
```

```python
def calculator(ops):
    def add(a,b):
        return a+b
    def substract(a,b):
        return a-b
    def multiply(a,b):
        return a*b
    
    if ops == "add":
        return add
    elif ops == "substract":
        return substract
    elif ops == "multiply":
        return multiply
    
```

```python
substract = calculator("substract")
substract(10,5)
```

```python
multiply = calculator("multiply")
multiply(4,5)

```

```python
num1 = int(input("enter first number:"))
num2 = int(input("enter the second number:"))

ops = input("enter operation (+,-,*,/):")
if ops == "+":
    res = num1 + num2
elif ops == "-":
    res = num1 - num2
elif ops == "*":
    res = num1 * num2
elif ops == "/":
    res = num1 / num2 
else:
    print("Invalid Operation")
    print("res:",res)


```
