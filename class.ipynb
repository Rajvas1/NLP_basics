```python
def show_dog(GT):
    print(f"Name: {GT['name']}, 'Breed':{GT['breed']}, Age: {GT['age']}")
```

```python
show_dog(GS)
```

```python
# Simple Defining Class
class dog:
    """ A dog class with constructor and methods"""
    def __init__(self,name,breed,age):
        self.name = name
        self.breed = breed
        self.age = age

    def bark(self):
        print(f" {self.name} bark: boww...boww")

    def show_info(self):
        print(f" dog's info: I am {self.name}, my breed{self.breed}, and I am {self.age} years old")

    def my_age(self):
        self.age +=1
        print(f" my name {self.name} now {self.age} years old")
```

```python
pug = dog("chers", "biscuit", 8)
```

```python
pug.bark()
```

```python
pug.show_info()
```

```python
pug.my_age()
```
